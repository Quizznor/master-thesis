```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hits = np.loadtxt("/cr/tempdata01/filip/QGSJET-II/lateral_density_function/hits.csv")
misses = np.loadtxt("/cr/tempdata01/filip/QGSJET-II/lateral_density_function/misses.csv")
```

```python
theta_bins = [0, 26, 38, 49, 60]
energy_bins = [16, 16.5, 17, 17.5, 18, 18.5, 19, 19.5]
sorted_container = [[[] for i in range(len(theta_bins) - 1) ] for i in range(len(energy_bins) - 1)]
```

#### Sort Misses

```python
miss_sorted = sorted_container

for miss in misses:

    spd, energy, theta, n_mu, n_e, n_gamma = miss
    e, t = energy_bins[-1], theta_bins[-1]
    energy_index, theta_index = -1, -1

    while energy < 10**e:

        energy_index -= 1
        e = energy_bins[energy_index]

    while theta < t:

        theta_index -= 1
        t = theta_bins[theta_index]

    miss_sorted[energy_index][theta_index].append(spd)
```

#### Sort Hits

```python
hits_sorted = sorted_container

for hit in hits:

    spd, energy, theta, n_mu, n_e, n_gamma = hit
    e, t = energy_bins[-1], theta_bins[-1]
    energy_index, theta_index = -1, -1

    while energy < 10**e:

        energy_index -= 1
        e = energy_bins[energy_index]

    while theta < t:

        theta_index -= 1
        t = theta_bins[theta_index]

    hits_sorted[energy_index][theta_index].append(spd)
```
