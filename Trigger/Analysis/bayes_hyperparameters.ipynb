```python
from TriggerStudyBinaries_v5 import BayesianClassifier
import numpy as np, matplotlib.pyplot as plt
```

```python
TestClass = BayesianClassifier(-249)

plt.plot(TestClass.bin_centers, TestClass.background)
plt.plot(TestClass.bin_centers, TestClass.signal)
plt.xscale("log")

# try to recover signal from signal + background
x_vals = TestClass.bin_centers
background = TestClass.background
signal = np.clip(TestClass.signal[:-1] - background[:-1], 0, 1) / np.sum(np.clip(TestClass.signal[:-1] - background[:-1], 0, 1))

plt.figure()
plt.plot(x_vals[:-1], signal)
plt.plot(x_vals, background)
plt.xscale("log")
```

```python
plt.plot(x_vals[:-1], signal / background[:-1])
plt.xscale("log"), plt.yscale("log")
plt.axhline(1, ls = "--", c = "k")
```

```python
print(x_vals[np.abs((signal / background[:-1]) - 1).argmin()])
```
