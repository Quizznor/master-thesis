```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_max = 8

class Line():

    def __init__(self, start, end, **kwargs):

        self.x_i, self.y_i = start
        self.x_f, self.y_f = end 

        ax1.plot([self.x_i, self.x_f], [self.y_i, self.y_f], c = kwargs.get("c", "k"), lw = 3)

n_rows = 7
split = 5

ax4 = plt.subplot2grid((n_rows, 3), (split, 2), rowspan = n_rows - split)
ax1 = plt.subplot2grid((n_rows, 3), (0, 0), colspan = 2, rowspan = split)
ax2 = plt.subplot2grid((n_rows, 3), (0, 2), sharey = ax1, rowspan = split)
ax3 = plt.subplot2grid((n_rows, 3), (split, 0), colspan = 2, rowspan = n_rows - split)

shift = +0.5
x_last = [0 - shift]

Line((x_last[0], 9), (x_last[0], 8), c = "steelblue")

for n in range(5):

    x_temp = []

    if n == 0:
        c_left, c_right = "orange", "steelblue"
    elif n == 1:
        c_left, c_right = "steelblue", "steelblue"
    elif n == 2:
        c_left, c_right = "orange", "steelblue"
    else:
        c_left = c_right = "k"

    for x in x_last:

        scaling_factor = (n + 1) * 0.9 if n > 0 else 0.5
        x_right, x_left = x + 1/scaling_factor, x - 1/scaling_factor
        x_temp.append(x_right), x_temp.append(x_left)

        if -2.5 <= x <= 0:
            Line((x, n_max - n), (x_right, n_max - n - 1), c = c_right)
            Line((x, n_max - n), (x_left, n_max - n - 1), c = c_left)            

        else:
            Line((x, n_max - n), (x_right, n_max - n - 1))
            Line((x, n_max - n), (x_left, n_max - n - 1))

    x_last = x_temp

    ax1.text(4.2, n_max - n + 0.1, f"n = {n}")
    ax1.axhline(n_max - n - 1, c = "gray", ls = "--", lw = 0.7)
    ax2.axhline(n_max - n - 1, c = "gray", ls = "--", lw = 0.7)
    ax2.plot([2**n, 2**n], [n_max - n, n_max - n+1] , c = "steelblue", lw = 2)
    ax2.plot([2**n, 2**(n+1)], [n_max - n, n_max - n], c = "steelblue", lw = 2)

ax1.spines['bottom'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax3.spines['top'].set_visible(False)
ax4.spines['top'].set_visible(False)

ax2.set_ylim(2.8, n_max + 0.8)
ax1.set_xlim(-5, 5)
ax2.set_xlim(0, 300)
ax4.set_xlim(0, 300)
ax3.set_xlim(-5, 5)
ax3.set_ylim(0.1)
ax4.set_ylim(0,1)

kwargs = dict(color='k', clip_on=False, lw = 0.7)
# ax1.plot((0.95 * 2.8, 1.05 * 2.8), (1.02, 1.08), **kwargs)
ax1.plot([0.98 * -5, 1.02 * -5], [1.01 * 2.8, 0.99 * 2.8], **kwargs)
ax1.plot([0.98 * 5, 1.02 * 5], [0.99 * 2.8, 1.01 * 2.8], **kwargs)
ax2.plot([-6, 6], [0.99 * 2.8, 1.01 * 2.8], **kwargs)
ax2.plot([294, 306], [0.99 * 2.8, 1.01 * 2.8], **kwargs)
ax3.plot([0.98 * 5, 1.02 * 5], [0.99, 1.01], **kwargs)
ax3.plot([0.98 * -5, 1.02 * -5], [1.01, 0.99], **kwargs)
ax4.plot([-6, 6], [0.99, 1.01], **kwargs)
ax4.plot([294, 306], [0.99, 1.01], **kwargs)

ax1.set_xticks([])
ax1.set_yticks([])
ax2.set_yticks([])
ax2.set_xticks([])
ax4.set_yticks([])
ax3.set_yticks([])
ax3.set_xticks([])
ax4.set_xlabel("N(n)")

ax1.text(4.2, n_max - 5 + 0.1, "n = 5")
ax1.axhline(n_max, c = "gray", ls = "--", lw = 0.7, zorder = 0)
ax2.axhline(n_max, c = "gray", ls = "--", lw = 0.7, zorder = 0)
ax2.plot([2**(n+1), 2**(n+1)], [n_max - ( n + 1 ), n_max - n] , c = "steelblue", lw = 2)

ax3.text(4.2, 0.5 + 0.05, "n = 8")
ax3.axhline(0.5, c = "gray", ls = "--", lw = 0.7, zorder = 0)
ax4.axhline(0.5, c = "gray", ls = "--", lw = 0.7, zorder = 0)

ax3.text(-shift - 0.265, 1,  "...", clip_on=False, size = 40, fontfamily = "serif")
ax4.plot([128, 256, 256, 0], [0.96, 0.96, 0.5, 0.5], c = "steelblue", lw = 2)

for _ in range(256):
    x0, y0 = np.random.normal(- shift, 1.6), 0.9
    x1, y1 = x0 + np.random.uniform(-0.3, 0.3), 0.5

    ax3.plot([x0, x1], [y0, y1], c = "k", lw = 0.5)
```
