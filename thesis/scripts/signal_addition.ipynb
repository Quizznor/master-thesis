```python
from Binaries import *

Events = EventGenerator("19_19.5", split = 1, real_background = True, station = "lo_qui_don", seed = 42, force_inject = 5)
Traces = Events.__getitem__(0)
example_trace = Traces[6]
example_trace.__plot__()
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable 

fig = plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))
divider = make_axes_locatable(ax1) 
ax2 = divider.append_axes("bottom", size="100%", pad=0.5)
ax3 = divider.append_axes("bottom", size="100%", pad=0.5)
ax4 = divider.append_axes("bottom", size="200%", pad=0.8)

# ax2 = plt.subplot2grid((4,1),(1,0), sharex = ax1, sharey = ax1)
# ax3 = plt.subplot2grid((4,1),(2,0), sharex = ax1, sharey = ax1)
# ax4 = plt.subplot2grid((4,1),(3,0), sharex = ax1)
pmts = [example_trace.pmt_1, example_trace.pmt_2, example_trace.pmt_3]

for i in range(3):

    ax1.plot(range(2048), example_trace.Signal[i])
    ax2.plot(range(2048), example_trace.Injected[i])
    ax3.plot(range(2048), example_trace.Baseline[i])
    ax4.plot(range(2048), pmts[i], label = f"PMT \#{i}")

for i in [example_trace.signal_start, example_trace.signal_end]:
    ax1.axvline(i, c = "red", ls = "--")
    ax4.axvline(i, c = "red", ls = "--")

for i in example_trace.injections_start + example_trace.injections_end:
    ax2.axvline(i, c = "gray", ls = "--")
    ax4.axvline(i, c = "gray", ls = "--")

ax1.set_xticks([],[])
ax2.set_xticks([],[])
ax3.set_xticks([],[])

ax1.legend(title="Shower simulation", loc = "upper right")
ax2.legend(title="Injected background", loc = "upper right")
ax3.legend(title="Random trace background", loc = "upper right")
ax2.set_ylabel("Component traces / $\mathrm{ADC}$")
ax4.set_xlabel("Bin / $8.3\,\mathrm{ns}$")
ax4.set_ylabel("Total / $\mathrm{VEM}_\mathrm{Peak}$", labelpad = 40)
ax4.legend()
plt.tight_layout()
```
