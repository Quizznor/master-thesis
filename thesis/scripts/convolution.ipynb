```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [40, 20]
```

```python
n_rows, n_cols = 6, 6



matrix = np.zeros((n_rows, n_cols))

# eyes
matrix[1, 1] = matrix[1, 4] = 1

# mouth
matrix[3, 1] = matrix[3, 4] = 1
matrix[4, 1:5] = 1

plt.matshow(matrix, cmap = "Greys")
```

```python
grid_resolution = 3


for plt_row in range(0, grid_resolution, grid_resolution // 3):
    for plt_col in range(0, grid_resolution, grid_resolution // 3):

        ax = plt.subplot2grid((grid_resolution, grid_resolution), (plt_col, plt_row), colspan = 1, rowspan = 1)
        ax.matshow(matrix, cmap = "Greys")

        for x, row in enumerate(matrix.T):
            for y, col in enumerate(row):
                ax.text(x - 0.29, y + 0.07, f"{col:.2f}", c = "k" if col < 0.5 else "w", fontsize = 15)

        # draw box
        ax.plot([plt_row - 0.5, plt_row + 1.5], [plt_col - 0.5, plt_col - 0.5], c = "r", lw = 5, clip_box = False) # top
        ax.plot([plt_row - 0.5, plt_row + 1.5], [plt_col + 1.5, plt_col + 1.5], c = "r", lw = 5, clip_box = False) # bot
        ax.plot([plt_row - 0.5, plt_row - 0.5], [plt_col - 0.5, plt_col + 1.5], c = "r", lw = 5, clip_box = False) # lft
        ax.plot([plt_row + 1.5, plt_row + 1.5], [plt_col - 0.5, plt_col + 1.5], c = "r", lw = 5, clip_box = False) # rgt

        ax.set_xticks([])
        ax.set_yticks([])

plt.tight_layout()
plt.savefig("/home/quizznor/git-repositories/master-thesis/thesis/imgs/convolution.png")
```
